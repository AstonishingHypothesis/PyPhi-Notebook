
So, now that we know all the basics, let's get started with IIT for real.

The first thing we need is more than one number. We need a lot of numbers (**data**). 

Since we are interested in IIT, these data will most likely constitute some kind of *measurement*. 

This could be a measurement of brain activity (we wil get to that). At its simplest, or most general and abstract, the data of interest constitute a state of a system, or parts of a system over time.

Since we start simple and only gradually get more complex, let's assume that the system that we are studying has only two states: ***On*** and ***Off***. Mathematically, these two states could be expressed quite simply as 1 and 0 (any number or assignment will do, of course, but 0 and 1 come in quite handy).

If we measure the system twice times over a certain period of time (every second or millisecond or whatever our choice of sampling rate would be), we end up with a single string of data, such as: 

1 1 

In other words, the first time we measure the system it is ON. Then, after waiting, we test again and find out the system is still ON.

OK. Let's do that in code:

```python
import numpy as np 
 
data = [1,1] 
 
```

If you are new to python, you might wonder at this point what `numpy` means. Well, python can do math and even complex math, but it needs help to do so. [numpy](https://numpy.org/) is a software package that, when imported, allows for linear algebra. 

[Linear algebra](https://en.wikipedia.org/wiki/Linear_algebra) is what we call the math surrounding vectors. And since a set of two numbers (0 1) can also be thought of as a [vector](https://en.wikipedia.org/wiki/Vector_(mathematics_and_physics)), we will make use of linear algebra from here on out.

Just for the fun of it, you can click the code below and look at our little data vector in a plot. Visualizing data before and after applying mathematical operations (such as those of IIT) to it can be very helpful to understand what happens at each step. We will thus make a habbit of periodically checking in on our data in the form of plots.

The simplest case of a vector are a pair of numbers (such as our 1 and 1). 

Thanks to Descartes, who first had the idea of a [coordinate system](https://en.wikipedia.org/wiki/Cartesian_coordinate_system), we now tend to think of a vector as an "arrow" within the coordinate system that points from the origin (the 0,0 point) to the coordinates set up by the vector (here: 1 and 1). 

So, the vector (1 1), can be visualized as the following below (just click and run the code - do not worry about understanding every line for now):

```python
import matplotlib.pyplot as plt
import math

head_length = 0.1
origin = [0,0]
dx = data[0] - origin[0]
dy = data[1] - origin[1]
vector = [dx,dy]
vector_magnitude = math.sqrt(dx**2+dy**2)
dx = dx / vector_magnitude
dy = dy / vector_magnitude
vector_magnitude = vector_magnitude - head_length

ax = plt.axes()
ax.arrow(origin[0], origin[1], vector_magnitude*dx, vector_magnitude*dy, head_width=0.1, head_length=0.1, fc='black', ec='black')
plt.xlim(0,3)
plt.ylim(0,3)
plt.grid()
plt.show()
plt.close()
```
